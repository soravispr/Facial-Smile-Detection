### Data Preparation Module

```python
import os, shutil

original_dir = './datasets/images/'
train_dir = './datasets/train'
validation_dir = './datasets/validation'
test_dir = './datasets/test'

## ---------------------- Not Smile -----------------------------------------
label = '0'
path, dirs, files = next(os.walk(original_dir + label))
file_count = len(files)
count = 1
train_num = int(file_count * 0.8)
validation_num = train_num + int(file_count * 0.2)
# test_num = int(file_count * 0.15)

for fname in files:
    if count <= train_num:
        src = original_dir + label + '/' + fname
        des = train_dir + '/' + label + '/' + fname 
        shutil.copyfile(src, des)
    elif count > train_num and count <= validation_num:
        src = original_dir + label + '/' + fname
        des = validation_dir + '/' + label + '/' + fname 
        shutil.copyfile(src, des)
#     else:
#         src = original_dir + label + '/' + fname
#         des = test_dir + '/' + label + '/' + fname 
#         shutil.copyfile(src, des)
        
    count +=1

## ---------------------- Smile -----------------------------------------
label = '1'
path, dirs, files = next(os.walk(original_dir + label))
file_count = len(files)
count = 1
train_num = int(file_count * 0.8)
validation_num = train_num + int(file_count * 0.2)
# test_num = int(file_count * 0.15)

for fname in files:
    if count <= train_num:
        src = original_dir + label + '/' + fname
        des = train_dir + '/' + label + '/' + fname 
        shutil.copyfile(src, des)
    elif count > train_num and count <= validation_num:
        src = original_dir + label + '/' + fname
        des = validation_dir + '/' + label + '/' + fname 
        shutil.copyfile(src, des)
#     else:
#         src = original_dir + label + '/' + fname
#         des = test_dir + '/' + label + '/' + fname 
#         shutil.copyfile(src, des)
        
    count +=1
```
